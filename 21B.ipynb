```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
```

```python
# Load the dataset into a DataFrame
df = pd.read_csv('wc-at.csv')
```

```python
# Extract the 'Waist' and 'AT' columns
waist_data = df['Waist']
at_data = df['AT']
```

```python
# Create histograms for Waist and AT
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(waist_data, kde=True)
plt.title('Waist Distribution Histogram')
plt.xlabel('Waist')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
sns.histplot(at_data, kde=True)
plt.title('AT Distribution Histogram')
plt.xlabel('AT')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Perform Shapiro-Wilk tests for Waist and AT
stat_waist, p_waist = shapiro(waist_data)
stat_at, p_at = shapiro(at_data)

# Check normality based on p-values
if p_waist > 0.05:
    print("Waist data appears to be normally distributed (p =", p_waist, ")")
else:
    print("Waist data does not appear to be normally distributed (p =", p_waist, ")")
print(" "*50)
if p_at > 0.05:
    print("AT data appears to be normally distributed (p =", p_at, ")")
else:
    print("AT data does not appear to be normally distributed (p =", p_at, ")")
```
