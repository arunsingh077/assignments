```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
```

```python
# Load your CSV data into a DataFrame
df = pd.read_csv('Cars.csv')
```

```python
# Extract the 'MPG' column
mpg_data = df['MPG']
```

```python
# Create a histogram
sns.histplot(mpg_data, kde=True)
plt.title('MPG Distribution Histogram')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()

# Perform the Shapiro-Wilk test
stat, p = shapiro(mpg_data)

if p > 0.05:
    print("MPG data appears to be normally distributed (p =", p, ")")
else:
    print("MPG data does not appear to be normally distributed (p =", p, ")")
```
